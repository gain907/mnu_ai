### 정답데이터가 한쪽으로 편중되어버리는 현상 발생
- 해결 : train_test_split

```python
# train_test_split
# 데이터를 랜덤하게 섞어서 학습데이터 , 테스트 데이터로 분류해주는 함수
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,
                 test_size=0.2, # 기본 : 0.25
                 random_state= 42) # 섞어주세요(아무런 숫자사용가능)
```

### 모델 만들기

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
iris_knn = KNeighborsClassifier(n_neighbors= 4)
iris_knn


```

```python
# 모델 학습
iris_knn.fit(X_train,y_train)
```

```python
# 모델 테스트
y_pred = iris_knn.predict(X_test)
```

```python
y_pred # 예측 답지
```

```python
y_test
```

```python
import numpy as np
# 모델 평가
# accuracy_score(예측답지,실제답지)
acc_score = accuracy_score(y_pred,y_test)
score = np.round(acc_score*100,2) # 소수점 2번째 자리까지만 보여주세요
print('score 값은 {}%'.format(score))
```

### 하이퍼 파라미터 튜닝
- k값 범위 : 1~71
- train,test 정확도를 비교

```python
for i in range(1,71):

  print(i)
```

```python
# 반복문 사용해서 하이퍼 파라미터 튜닝 후 결과를 한번에 확인
# 결과를 저장할 리스트 변수 2개 생성
# 변수 생성
train_acc = [] # 리스트 변수
test_acc = []  # 리스트 변수
# k값(파라미터 값) 설정(1~71)
n_neighbors_set = range(1,71) # 1~70

# for i in range(1,71): == for i in n_neighbors_set
for n in n_neighbors_set:
  # 모델 생성
  knn_model = KNeighborsClassifier(n_neighbors=n)
  # 모델 학습
  knn_model.fit(X_train,y_train)
  # score값
  train_score = knn_model.score(X_train,y_train)
  test_score = knn_model.score(X_test,y_test)
  # score값을 리스트에 저장
  train_acc.append(train_score)   # .append 리스트에서 값을 추가할때
  test_acc.append(test_score)     # .append 리스트에서 값을 추가할때


```

```python
train_acc
```

```python
test_acc
```

```python
# 그래프로 결과 확인
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
# train_acc 값을 그래프로 보여주세요
plt.plot(n_neighbors_set,train_acc, label='Train_acc', marker='o')
# test_acc 값을 그래프로 보여주세요
plt.plot(n_neighbors_set,test_acc, label='Test_acc', marker='o')
plt.xticks(range(1,71,2))
plt.grid() # 격자 형태로 그래프를 그려주세요
plt.legend()
plt.xlabel('n_neighbors_number')
plt.ylabel('Acc Score')
plt.show() # 그래프를 출력해 주세요

```
